# Chest image  using imited adaptive histogram equalization (CLAHE)

## chest image problem

Chest radiography is a common diagnostic imaging test, which contains an enormous amount of information about a patient
enhancement of the images is indispensable in order to improve visibility of the details. This paper aims at improving radiograph parameters such as contrast, sharpness, noise level, and brightness to enhance chest radiographs.Doing so helps doctors and radiologists better interpret the scans and make an accurate diagnosis.
I notic that it is poor image contrast and nonuniform image intensity
Histogram equalization:processing method to stretch the histogram of the image intensity evenly according to pixel intensity probability.However, HE is not able to avoid high peaks (i.e., clusters of image intensity) in the histogram; therefore cannot CLAHE:enhance the contrast between pixels with the peaks, within a small range of image intensity
has been developed to overcome such limitations by processing the image histogram in blocks, limiting the intensity dynamic range, and then clipping and redistributing the gray peaks.
so,here we use CLAHE algorithim to enhance the image.

```python
img = cv2.imread('img\chest.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #convert image to RGB
# creating a Histograms Equalization
# of a image using cv2.equalizeHist()
blur = cv2.GaussianBlur(img,(9,9),sigmaX=0, sigmaY=0)
equ = cv2.equalizeHist(blur)

# The declaration of Adaptive Histrogram Equalization - regions not pixels 
# clipLimit -> Threshold for contrast limiting
# CLAHE is a variant of Adaptive histogram equalization (AHE) which takes care of over-amplification of the contrast. 
clahe = cv2.createCLAHE(clipLimit = 150) 
final_img = clahe.apply(img)
ret, thresh = cv2.threshold(final_img, 160, 255, cv2.THRESH_BINARY)


plt.figure(figsize=(30,20))
plt.subplot(231), plot_imageGray(img, 'Original image')
plt.subplot(232), plot_imageGray(final_img, 'Adaptive Equlized Image')

plt.show()
```

# bone friction image

```python

image = cv2.imread('img/br.jpg')
orig_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

height, width = img_gray.shape

blur = cv2.medianBlur(img_gray, 3)

ret, thresh = cv2.threshold(blur, 160, 255, cv2.THRESH_BINARY)

contours, hierarchy = cv2.findContours(image=thresh, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)[-2:]

image_copy = orig_image.copy()
contourB = np.zeros(img_gray.shape) 
cv2.drawContours(image=thresh, contours=contours, contourIdx=-1, color=(255, 255, 255), thickness=5, lineType=cv2.LINE_AA)[-2:]
for ind, contour in enumerate(contours):
    
    area = cv2.contourArea(contour)
    valume = (4 / 3) * np.pi * np.sqrt(area / np.pi)**3
    circimg = image.copy()
    # Plot a red circle on the centroid of the round objects
    round_thresh = 100000

    if valume > round_thresh:
        # find center of the blob object
        moments = cv2.moments(contour)
        cx = int(moments['m10'] / moments['m00'])
        cy = int(moments['m01'] / moments['m00'])
        #put circle in the cent or the image (in this case it will shown as dot)
        cv2.circle(orig_image, (cx,cy), 2, (255, 0, 0), 2, cv2.LINE_AA)

#Plot the rest of the image for viewing
plt.figure(figsize=(20,20))
plt.subplot(221),plot_image(circimg, 'Original Image')
plt.subplot(222),plot_imageGray(thresh, 'Contours')
plt.subplot(223),plot_image(orig_image, 'Circle Location')

plt.show()
```

# back image

```python
back= cv2.imread("img/x.jpg", cv2.IMREAD_GRAYSCALE)
back = cv2.medianBlur(back, 3)
element = cv2.getStructuringElement(shape = cv2.MORPH_RECT, ksize=(11,11))
erosion_dst = cv2.erode(back, element)
result = back - erosion_dst

ret, thresh = cv2.threshold(result, 25, 255, cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(image=thresh, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)[-2:]
cv2.drawContours(image=thresh, contours=contours, contourIdx=-1, color=(255, 255, 255), thickness=10, lineType=cv2.LINE_AA)[-2:]


element = cv2.getStructuringElement(shape = cv2.MORPH_CROSS, ksize=(5,5))
thresh = cv2.erode(thresh, element)
plt.figure(figsize=(20,20))
plt.subplot(121), plot_imageGray(back,'Orginal image')
plt.subplot(122), plot_imageGray(thresh,'Boundary Extraction')
plt.show()
```

Median bluer:
This is a non-linear filtering technique. As clear from the name, this takes a median of all the pixels under the kernel area and replaces the central element with this median value. This is quite effective in reducing a certain type of noise.

Contrast Stretching:
It is one of the piecewise linear function. Contrast Stretching increases the dynamic range of the grey level in the image being processed. The Points (r1, s1) and (r2, s2) control the shape of the transformation. The selection of control points depends upon the types of image and varies from one image to another image. If r1 = s1 and r2 = s2 then the transformation is linear and this doesn’t affect the image.
