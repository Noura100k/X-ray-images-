# Bone friction image 


## The image problem:

* introduction: 
Knee osteoarthritis is defined by degeneration of the knee’s articular cartilage—the flexible, slippery material that normally protects bones from joint friction and impact. 


* The problems in image: 
we try to help doctors when they see the x-ray image for knee injured or not with knee osteoarthritis easily.After applying DIP,if the knee region in image  has 2 dots , it means the knee is normal but if the knee region in image has 1 dot, it means the knee has Knee osteoarthritis .



## The using method to solve the problem:
Feature

-  A feature is a piece of information which is relevant for solving the computational task related to a certain application. Features may be specific structures in the image such as points, edges or objects. Features may also be the result of a general neighborhood operation or feature detection applied to the image. 
- Main Component 

 - Detection: Identify the Interest Point
 - Description: The local appearance around each feature point is described in some way that is (ideally) invariant under changes in illumination, translation, scale, and in-plane rotation. We typically end up with a descriptor vector for each feature point.

contour detection
- A contour is a closed curve joining all the continuous points having some color or intensity, they represent the shapes of objects found in an image. Contour detection is a useful technique for shape analysis and object detection and recognition.
- It provides two simple functions:
 -  findContours(): Finds contours in a binary image.
 - drawContours(): Draws contours outlines or filled contours.



## How this method help to solve the probelm:
l will illustrate how the using method helps us to find the solution by explaining the code as following:
*  load the image
*  Converting it to RGB and then grayscale
*  smoothing the image to remove any nois.
*  we need to create a binary image, which means each pixel of the image is either black or white. This is a necessity in OpenCV, finding contours is like finding a white object from a black background, objects to be found should be white and the background should be black using threshold.the value of threshold is 160.So, all values above it consider as forground and below it as background.
*  Detect contours by using findContours. l pass a threshold image as an argument of function.After finding all contours in image, we use draw contours.
* Equivalent Spherical Volume: Is the Volume of the sphere that has a diameter of the Equivalent Spherical Diameter.we use (4/3) x PI x sqrt(Area/PI)^3 to calculate the valume.we make a loop to go to all points in contour and  if the valume > 100000  ,it draw a dot.

the algorithm draws a dot for each object that has a counter. if the objects are near too much, the algorithm considers it as one region implying the patient has Knee osteoarthritis. So, we draw one dot. If the objects are not close to much, there is some space between them. Then, the algorithm considers them as distinct objects implying that the patient does not have Knee osteoarthritis. So, we draw 2 dots in our case.
## the challenges:
* To make the algorithm understand that we want to distinguish two images that are infected with Knee osteoarthritis and the other is not.Discrimination Whether the bone above is separate from the bone below, or are they close to much as one bone.
* determine the best value(round_thresh in the code) that dtermine if the dot should be drawn or not.


```python

image = cv2.imread('img/br.jpg')#read the image
orig_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)#convert image to RGB
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)#convert image to GRAY

height, width = img_gray.shape # store the width and height for the image

blur = cv2.medianBlur(img_gray, 3)# apply median blure filter on the image

ret, thresh = cv2.threshold(blur, 160, 255, cv2.THRESH_BINARY)#convert it to binary image

contours, hierarchy = cv2.findContours(image=thresh, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)[-2:]
#find the contours that present in the image for (forground)
image_copy = orig_image.copy()#make shallow copy
contourB = np.zeros(img_gray.shape) #creat an image with all zero as the same size of orginal(black image)
cv2.drawContours(image=thresh, contours=contours, contourIdx=-1, color=(255, 255, 255), thickness=5, lineType=cv2.LINE_AA)[-2:]
#strating to draw the countours
for ind, contour in enumerate(contours):
    
    area = cv2.contourArea(contour)#finding area of countor
    valume = (4 / 3) * np.pi * np.sqrt(area / np.pi)**3 #finding valumw of countor
    circimg = image.copy()#make shallow copy
    # Plot a yellow circle on the centroid of the round objects
    round_thresh = 100000 # value of round threshold

    if valume > round_thresh:
        '''Image moments help you to calculate some features
           like center of mass of the object, area of the object '''
        moments = cv2.moments(contour)#The function cv.moments() gives a dictionary of all moment values calculated
        cx = int(moments['m10'] / moments['m00'])#get x value to draw the circle at the center
        cy = int(moments['m01'] / moments['m00'])#get y value to draw the circle at the center
        cv2.circle(orig_image, (cx,cy), 2, (255, 255, 0), 2, cv2.LINE_AA)#function to draw a circle

#Plot the rest of the image for viewing
plt.figure(figsize=(20,20))
plt.subplot(221),plot_image(circimg, 'Original Image')
plt.subplot(222),plot_imageGray(thresh, 'Contours')
plt.subplot(223),plot_image(orig_image, 'Detecting Bone friction')

plt.show()
```

# Chest image  


## The image problem:

* Introduction: 
Chest radiography is a common diagnostic imaging test, which contains an enormous amount of information about a patient
enhancement of the images is indispensable in order to improve visibility of the details. This paper aims at improving radiograph parameters such as contrast, sharpness, noise level, and brightness to enhance chest radiographs.Doing so helps doctors and radiologists better interpret the scans and make an accurate diagnosis.

* The problems in the image: 
I notic that it is poor image contrast and nonuniform image intensity




## The using method to solve the problem:

* Histogram equalization:processing method to stretch the histogram of the image intensity evenly according to pixel intensity probability.However, HE is not able to avoid high peaks (i.e., clusters of image intensity) in the histogram.
* CLAHE:enhance the contrast between pixels with the peaks, within a small range of image intensity
has been developed to overcome such limitations by processing the image histogram in blocks, limiting the intensity dynamic range, and then clipping and redistributing the gray peaks.
so,here we use CLAHE algorithim to enhance the image.



## How this method help to solve the probelm:
l will illustrate how the using method helps us to find the solution by explaining the code as following:
*  load the image
*  Converting it to RGB and then grayscale
*   cv2.createCLAHE function with value of threshold equal to 150 for contrast limiting.If noise is there, it will be amplified. To avoid this, contrast limiting is applied

 image is divided into small blocks called "tiles".Then each of these blocks are histogram equalized as usual. So in a small area, histogram would confine to a small region (unless there is noise). If noise is there, it will be amplified. To avoid this, contrast limiting is applied. If any histogram bin is above the specified contrast limit those pixels are clipped and distributed uniformly to other bins before applying histogram equalization. After equalization, to remove artifacts in tile borders, bilinear interpolation is applied.

## the challenges:
* make the image as uniform image intensity
* trying diffrent values for clipLimit


```python
img = cv2.imread('img\chest.jpg')#read the image
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #convert image to GRAY
clahe = cv2.createCLAHE(clipLimit = 150) #threshold for contrast limiting
final_img = clahe.apply(img)#apply clahe in image
plt.figure(figsize=(30,20))
plt.subplot(231), plot_imageGray(img, 'Original image')
plt.subplot(232), plot_imageGray(final_img, 'Adaptive Equlized Image')

plt.show()
```

# Hand image 


## The image problem:

X-ray of hands can help the doctors to detect any injuries in the radiocarpal and distal radioulnar joints, the carpals, metacarpals, and phalanges.
The edges hand x-ray very important as the doctors can see clearly the bones ,whether it is tiny or visible bone.

The edges in the image are not clear as much as the doctors need, the DIP help in fixing that by using sobel edge detection (Sobel filter).





## The using method to solve the problem:
Gaussian Blur:It is an essential part of many image processing algorithms. It consider as a non-uniform noise reduction low-pass filter (LP filter). It is also consider as a spatial filter that works by convolving the input image with a kernel. The visual effect of this filter is a smooth blurry image. This filter works by taking a pixel and calculating a value (similar to the mean, but with more bias in the middle). The filter is constructed based on the normal distribution, which is shaped like a bell curve. The idea is that pixels closer to the center pixel have a larger weight and those further a way a smaller weight. The math equations below show how to calculate the proper weights of the kernel.

Gaussian Blur=1/(2π σ) e^(-(x^2+y^2)/(2σ^2 ))
where σ is The standard deviation, in other words the average distance from the mean, determines how wide or narrow the bell curve is.

We can summarize some of the Gaussian’s filter features:
* Since this is a separable filter, we can save computing power
* The use of “weighted” masks makes it better for detecting edges than some uniform blurring filters
* Multiple iterations with a given filter size have the same blur effect as the larger one
* Useful as a pre-processing step for image size reduction

We can apply Gauusain Filter with the help of OpenCV library by simply calling the method GaussianBlur(source_image, kernel_size, sigmaX). Where,
* source_image is the image that is to be blurred using Gaussian Blur() function.
* kernel_size is the matrix representing the size of the kernel.
* sigmaX is a variable representing the standard deviation of Gaussian kernel in X direction and it is of type double.

Sobel Operator: It is an edges detector filter which tries to find the regions in an image where there is a sharp change in intensity or a sharp change in color, a high value indicates a steep change and a low value indicates a shallow change. It works by detecting a change in the gradient of the image in both the horizontal and vertical directions. To do this two convolution filters are applied to the original image, the results of these convolution filters are then combined to determine the magnitude of the gradient. It can be applied with the help of OpenCV libraray using the following method. 
cv2.Sobel(source_image, data type, dx, dy). Where,
* source_image to detect edge 
* output data type like cv2.CV_16S or cv2.CV_64F
* dx and dy specify whether Sobel-x or Sobel -y is to be used

Combining two image: OpenCV library in Python helps us in adding two images. These images can be added and used together for further analysis. A way of adding images is through blending them so that they both can be visible together. We can give different weights to different images, which can make the image transparent or translucent according to the weight added.The images can be added by using the following method.
cv2.addWeighted(source1, Alpha, source2, Beta, Gamma)
* source1 is the first image. 
* Alpha is the weighte first image (Alpha will be multiplied with source1).  
* source2 is the second image. 
* Beta is the weighte for second image (Beta will be multiplied with source2). 
* Gamma is the value that will be added to the multiplication result.

Algorithm to solve the problem:
* Read the image.
* Apply Gaussian Blur on the image. 
* Detect the edges of image only on Y axis using Sobel edge detector. 
* Get the Absolut value of Y detected edge on Y axis.
* Combine both Gaussian Blurred image and Y using addWeighted function.
* Plot the modified image. 

## How this method helps to solve the problem:
* Gaussian blur help to reduce the noise from the image to enhance the detection. We applied Gaussian filter with kernel size equal to 5X5 and segmaX as 0 
* Sobel operator help to detect the edges of the blurred image on one direction which is on Y axis 
* addWeighted function combined both detected edges on Y-axis and the blurred image. Also The selection of Alpha as 1.9 and Beta as 1.7 Gama as 3 increased the lightness of the image.



## the challenges:
* The only challenges in this image was to select good value for Alpha, Beta, and Gama

```python
image = cv2.imread('img/hand.jpg')#read the image
gussian_blur = cv2.GaussianBlur(image,(5,5),0) # Applies the Gaussian filter with kernel 5x5
y = cv2.Sobel(gussian_blur, cv2.CV_16S, 0, 1)  #Edge detection on Y axis
# Get the absolut value of Y detected edge on Y axis
absY = cv2.convertScaleAbs(y)
#Combine both Y and blured image
Sobel = cv2.addWeighted(gussian_blur,1.9, absY,1.7, 3) 
#ploting the image 
plt.figure(figsize=(20,20))
plt.subplot(121), plot_image(image, 'Original image') 
plt.subplot(122), plot_image(Sobel,'Sobel operation')
plt.show()
```


# Knee image 


## The image problem:

 * introduction: A knee radiographs is a common emergency procedure that can be performed quickly and easily. Having the proper knowledge of the anatomy and review strategies can help ensure that the diagnosis is accurate.

* the problem in image:Contarst of colors in the image is a main problem in x-ray images.With DIP,we can enhance the color contrast of images either graylevel images or RGB images.

## The using method to solve the problem:
* Gaussain Blur: The Gaussian blur feature is obtained by smoothing an image using a Gaussian function to reduce the noise level.Gaussian Blur=1/(2π σ) e^(-(x^2+y^2)/(2σ^2 )) where σ is The standard deviation, in other words the average distance from the mean, determines how wide or narrow the bell curve is.
        
* Contrast Stretching:
It is one of the piecewise linear function. Contrast Stretching increases the dynamic range of the grey level in the image being processed. The Points (r1, s1) and (r2, s2) control the shape of the transformation. The selection of control points depends upon the types of image and varies from one image to another image. If r1 = s1 and r2 = s2 then the transformation is linear and this doesn’t affect the image. 



## How this method help to solve the probelm:
 * The selection of control points depends upon the types of image and varies from one image to another image. If r1 = s1 and r2 = s2 then the transformation is linear and this doesn’t affect the image. In other case we can calculate the intensity of output pixel, provided intensity of input pixel is x, as follows
* for 0 <= x <= r1
    output = s1 / r1 * x
* for r1 < x <= r2
    output = ((s2 – s1)/(r2 – r1))*(x – r1) + s1
* for r2 < x <= L – 1
    output = ((L-1 – s2)/(L-1 – r2))*(x – r2) + s2


## the challenges:
The only challenges in this image was to select good value for r1, r2, s1 and s2

# Knee image problem

```python
# read image
image = cv2.imread('img/knee.jpg')#read the image
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)#convert to Gray 
gussian_blur = cv2.GaussianBlur(image,(11,11),0)#bluring the image using Gaussian to reduce any noise
def pixelVal(pix, r1, s1, r2, s2): #function to return the pixels after doing calculations
    if (pix >= 0 and pix <= r1): 
        return (s1 / r1) * pix 
    elif (r1 < pix and pix <= r2): 
        return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1 
    else: 
        return ((255 - s2)/(255 - r2)) * (pix - r2) + s2 
# Define parameters.
r1 = 1 # r1 we use it to make the dark region darker
s1 = 0 #the best value for this one is zero (0 dark intesity)
r2 = 100 # r2 make the regions of image lighter
s2 = 255 #the best value for this one is 255 (255 white intesity)
  
pixelVal_vec = np.vectorize(pixelVal)
result = pixelVal_vec(gussian_blur, r1, s1, r2, s2)
'''
vectorized function which takes a nested sequence
of objects or numpy arrays as inputs and returns 
a single numpy array or a tuple of numpy arrays.
'''

plt.figure(figsize=(20,20))
plt.subplot(121), plot_imageGray(image, 'Original image') 
plt.subplot(122), plot_imageGray(result ,'Contrast streaching')
plt.show()
```
